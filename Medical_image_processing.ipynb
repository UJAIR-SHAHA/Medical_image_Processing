##MEDICAL IMAGE 1
## X-ray of hand

```python
# Load the medical image
image_path = '/content/hand x ray.jpg'
image = cv2.imread(image_path)
```

```python
# Check if the image was loaded successfully
if image is None:
    print('Error: Unable to load the image.')
else:
    # Convert the image to grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to segment the image
    _, thresholded_image = cv2.threshold(grayscale_image, 128, 255, cv2.THRESH_BINARY)

    # Find contours in the thresholded image
    contours, _ = cv2.findContours(thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw the contours on the original image
    result_image = image.copy()
    cv2.drawContours(result_image, contours, -1, (0, 0, 255), 2)

```

```python
import matplotlib.pyplot as plt
```

```python
# Display the original image and the processed image
plt.figure(figsize=(12, 6))
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(122), plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB)), plt.title('Processed Image')
plt.show()
```

Bone Density and Integrity:

Evaluate bone density and integrity by examining the clarity of bone structures in the X-ray.
Identify fractures, dislocations, or abnormalities in the hand's bones.

```python
# Display the grayscale X-ray image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), cmap='gray')
plt.title('Grayscale X-ray Image')
plt.axis('off')
plt.show()
```

```python
# Apply Gaussian blur to reduce noise (adjust the kernel size as needed)
blurred_image = cv2.GaussianBlur(grayscale_image, (5, 5), 0)

# Display the Blurred image
plt.imshow(blurred_image, cmap='gray')
plt.title('blurred_image')
plt.axis('off')
plt.show()
```

```python
# Load the medical image
image_path2 = '/content/MRI_of_Human_Brain.jpg'
image2 = cv2.imread(image_path2)
```

```python
# Check if the image was loaded successfully
if image2 is None:
    print('Error: Unable to load the image.')
else:
    # Convert the image to grayscale
    grayscale_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to segment the image
    _, thresholded_image2 = cv2.threshold(grayscale_image2, 128, 255, cv2.THRESH_BINARY)

    # Find contours in the thresholded image
    contours, _ = cv2.findContours(thresholded_image2, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw the contours on the original image
    result_image2 = image2.copy()
    cv2.drawContours(result_image2, contours, -1, (0, 255, 0), 2)

```

```python
# Display the Blurred image
plt.imshow(grayscale_image2, cmap='gray')
plt.title('grayscale_image2')
plt.axis('off')
plt.show()
```

```python
# Apply edge detection to identify soft tissue edges
edges2 = cv2.Canny(blurred_image2, 15, 20)
# Display the edges
plt.imshow(edges2, cmap='gray')
plt.title('edges')
plt.axis('off')
plt.show()
```

```python
# Apply thresholding to segment brain tissues
_, segmented_image = cv2.threshold(grayscale_image2, 60, 255, cv2.THRESH_BINARY)

# Display
plt.imshow(segmented_image, cmap='gray')
plt.title('segmented_image')
plt.axis('off')
plt.show()
```

```python
# Apply thresholding to detect tumors
_, binary_image = cv2.threshold(grayscale_image2, 120, 190, cv2.THRESH_BINARY)


# Display
plt.imshow(binary_image, cmap='gray')
plt.title('binary_image')
plt.axis('off')
plt.show()
```

```python
only_brain_image = cv2.subtract(segmented_image,binary_image)

# Display
plt.imshow(only_brain_image, cmap="bone")
plt.title('only_brain_image')
plt.axis('off')
plt.show()
```
